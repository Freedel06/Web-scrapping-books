```python
#importation des différents modules 
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url='http://books.toscrape.com/'
```

```python
#recuperer les informations du site
r=requests.get(url)
```

```python
#recuprerer le fichier htlm du site
soup=BeautifulSoup(r.text)
lis=soup.find_all('a')
```

```python
#Mettre dans une liste tout les liens permettant d'avaoir accès 
#aux différents types de livres 
links=[]
for link in lis:
    e=link.get('href')
    links.append('http://books.toscrape.com/'+e)
```

```python
#selectionner mes les 5 catégories de livres à scrapper
#les catégories selectionner sont: Travel, classic, mystery, historical,sequential
interest=links[3:8] 
interest
```

```python
#Creer une liste vide pour chaque catégorie de livre 
mystery=[]
historical=[]
sequential=[]
classic=[]
travel=[]
```

```python
#classer à partir des urls  de chaque livres  receuillies sur le site , les différents livres dans leurs catégories respectives 
#Etant donné que toutes le catégories de livres n'ont pas les mêmes nombres de pages, nous classerons d'abord les livres dont 
#les catégories ont plusieurs pages(mystery,historical,sequential)
for i in range(1,5):
    url_2='http://books.toscrape.com/catalogue/category/books/mystery_3'+'/page-'+str(i)+'.html'
    url_3='http://books.toscrape.com/catalogue/category/books/historical-fiction_4'+'/page-'+str(i)+'.html'
    url_4='http://books.toscrape.com/catalogue/category/books/sequential-art_5'+'/page-'+str(i)+'.html'
    r2=requests.get(url_2)
    r3=requests.get(url_3)
    r4=requests.get(url_4)
    s2=BeautifulSoup(r2.text)
    s3=BeautifulSoup(r3.text)
    s4=BeautifulSoup(r4.text)
    p2=s2.find_all('h3')
    p3=s3.find_all('h3')
    p4=s4.find_all('h3')
    for h3 in p2:
        c=h3.find('a')
        l=c.get('href')
        mystery.append("https://books.toscrape.com/catalogue/"+l.replace('../../../', ''))
    for h3 in p3:
        b=h3.find('a')
        k=b.get('href')
        historical.append("https://books.toscrape.com/catalogue/"+k.replace('../../../', ''))
    for h3 in p4:
        a=h3.find('a')
        r=a.get('href')
        sequential.append("https://books.toscrape.com/catalogue/"+r.replace('../../../', '') )
```

```python
#Classement des livres dont les catégories ont une seul page 
url_1='http://books.toscrape.com/catalogue/category/books/travel_2/index.html'
url_5='http://books.toscrape.com/catalogue/category/books/classics_6/index.html'
r=interest[0]
q=url_5
t=requests.get(r)
q1=requests.get(q)
q2=BeautifulSoup(q1.text)
t1=BeautifulSoup(t.text)
p1=t1.find_all('h3')
p2=q2.find_all('h3')
for h3 in p1:
    e=h3.find('a')
    oo=e.get('href')
    travel.append("https://books.toscrape.com/catalogue/"+oo.replace('../../../', ''))
for h3 in p2:
    q4=h3.find('a')
    ee=q4.get('href')
    classic.append("https://books.toscrape.com/catalogue/"+ee.replace('../../../', ''))
del classic[-1]
```
